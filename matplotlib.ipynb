MatPlotLib:

```python
# importing matplotlib library

import matplotlib.pyplot as plt
```

```python
# importing numpy library

import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
z = np.cos(x)
```

Plotting a Data

```python
# sine wave

plt.plot(x, y)
plt.show()
```

```python
# # consine wave

# plt.plot(x, z)
# plt.show()
```

```python
#  adding title x -- axis and y -- axis labels --> sine wave

plt.plot(x, y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#  adding title x -- axis and y -- axis labels --> cosine wave

plt.plot(x, z)
plt.xlabel('angle')
plt.ylabel('cosine value')
plt.title('Cosine Wave')
plt.show()
```

```python
# parabola
x = np.linspace(-10, 10, 20)
y = x ** 2
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y, 'r+')
plt.show()
```

```python
plt.plot(x, y, 'b.')
plt.show()
```

```python
plt.plot(x, y, 'rx')
plt.show()
```

```python
x = np.linspace(-5, 5, 50)
plt.plot(x, np.sin(x), 'g-')
plt.plot(x, np.cos(x), 'r--')
plt.show()
```

Bar Plot  ** Important Plot

```python
figure = plt.figure()
ax = figure.add_axes([0.2, 0.2, 0.6, 0.6])
languages = ['eng', 'french', 'germen', 'spanish', 'Latin']
people = [100, 50, 150, 80, 40]

ax.bar(languages, people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.title('People Speake In')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
languages = ['eng', 'french', 'germen', 'spanish', 'Latin']
people = [100, 50, 150, 80, 40]

ax.pie(people, labels=languages, autopct='%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0, 0, 1, 1])

ax.scatter(x , y, color='g')
ax.scatter(x, z, color='b')

plt.show()
```

3D Scatter Plot

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)

ax.scatter(x, y, z, c=z, cmap='Blues')
plt.show()
```
