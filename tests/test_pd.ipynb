```python
import pandas as pd
import numpy as np

# 1. Create DataFrame
data = {
    'Name': ['Anu', 'Ravi', 'Fathima', 'John', 'Sneha'],
    'Roll No': [101, 102, 103, 104, 105],
    'English': [78, 45, 88, 34, 92],
    'Maths': [82, 90, 74, 67, 59],
    'Science': [91, 87, 77, 45, 33],
    'Social': [85, 56, 82, 49, 41],
    'Hindi': [79, 69, 90, 38, 35]
}

df = pd.DataFrame(data)

# 2. Total & Average
df['Total'] = df[['English', 'Maths', 'Science', 'Social', 'Hindi']].sum(axis=1)
df['Average'] = df['Total'] / 5

# 3. Result (Pass if all subjects ‚â• 35)
df['Result'] = df[['English', 'Maths', 'Science', 'Social', 'Hindi']].apply(lambda row: 'Pass' if all(row >= 35) else 'Fail', axis=1)

# 4. Grade
def assign_grade(avg, result):
    if result == 'Fail':
        return 'F'
    elif avg >= 90:
        return 'A+'
    elif avg >= 80:
        return 'A'
    elif avg >= 70:
        return 'B'
    elif avg >= 60:
        return 'C'
    elif avg >= 50:
        return 'D'
    else:
        return 'E'

df['Grade'] = df.apply(lambda row: assign_grade(row['Average'], row['Result']), axis=1)

# 5. Students who passed
passed = df[df['Result'] == 'Pass']

# 6. Subject-wise average (using numpy)
subject_cols = ['English', 'Maths', 'Science', 'Social', 'Hindi']
subject_averages = np.round(df[subject_cols].mean(), 2)

# 7. Topper
topper = df[df['Total'] == df['Total'].max()]['Name'].values[0]

# 8. Sorted by Average
sorted_df = df.sort_values(by='Average', ascending=False)

# 9. Save to Excel
df.to_excel("students_report.xlsx", index=False)

# üñ®Ô∏è Print summary
print("üéì Full Report:")
print(df)
print("\n‚úÖ Students Passed:")
print(passed[['Name', 'Result', 'Grade']])
print("\nüìä Subject Averages:")
print(subject_averages)
print(f"\nüèÜ Topper: {topper}")
```
